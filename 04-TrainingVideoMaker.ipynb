```python
import cv2
import os
import math
from matplotlib import pyplot as plt

rootfolder = 'results-20240922-071227'

vertical = False

size = (64*16*2,64*16*2)

if size[1]>size[0]:
    vertical = False
    size = (size[1],size[0])

vid_fname = os.path.join(rootfolder,'gans_training.mp4')
```

```python
real_images = []
gen_images = []
for folder in os.listdir(rootfolder):
    if folder.isnumeric():
        real_images.append(os.path.join(rootfolder, folder,"real.png"))
        gen_images.append(os.path.join(rootfolder, folder,"generated.png"))
```

```python
real_images,gen_images
```

```python
#cap = cv2.VideoCapture(0)
out = cv2.VideoWriter(vid_fname,cv2.VideoWriter_fourcc(*'mp4v'), 1, size)
```

```python
plt.imshow(cv2.cvtColor(cv2.imread(real_images[0]), cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(cv2.imread(gen_images[0]), cv2.COLOR_BGR2RGB))
```

```python
frames = []
for real,gen in zip(real_images,gen_images):
    #print(real)
    if vertical:
        im_v = cv2.resize(cv2.vconcat([cv2.imread(real), cv2.imread(gen)]),size)
    else:
        im_v = cv2.resize(cv2.hconcat([cv2.imread(real), cv2.imread(gen)]),size)
    frames.append(im_v)
for f in frames:
    out.write(f)
```
